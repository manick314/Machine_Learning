```python
import pandas as pd
import sklearn
```

```python
dataset=pd.read_csv('cancer_reg.csv', encoding='ISO-8859-1')
dataset.head()
```

```python
pd.set_option('display.max_columns',None)
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.head()
```

```python
cat=dataset.select_dtypes(include='object').columns
```

```python
cat
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_bin=LabelEncoder()
le_geo=LabelEncoder()
```

```python
dataset['binnedInc']=le_bin.fit_transform(dataset['binnedInc'])
dataset['Geography']=le_geo.fit_transform(dataset['Geography'])
```

```python
X=dataset.drop('TARGET_deathRate',axis=1)
y=dataset['TARGET_deathRate']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=48)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_predict=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_predict,y_test)
```

```python
from sklearn.feature_selection import VarianceThreshold
var=VarianceThreshold()
X_train_reduce=var.fit_transform(X_train)
X_test_reduce=var.transform(X_test)
```

```python
selected_features=X_train.columns[var.get_support()]
```

```python
selected_features
```

```python
X_train_reduce_df=pd.DataFrame(X_train_reduce,columns=selected_features)
```

```python
X_test_reduce_df=pd.DataFrame(X_test_reduce,columns=selected_features)
```

```python
lr2=LinearRegression()
```

```python
lr2.fit(X_train_reduce_df,y_train)
```

```python
y_predict_1=lr2.predict(X_test_reduce_df)
```

```python
r2_score(y_predict_1,y_test)
```
